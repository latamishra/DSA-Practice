```python
# 4ï¸âƒ£ Move Zeroes
# Problem:
# Move all 0s to the end while keeping the order of non-zero elements.
# Key idea: Two pointers.
# Example
# nums = [0,1,0,3,12]
# Output â†’ [1,3,12,0,0]

def moveZeros(nums):
    for i in range(len(nums)):
        if nums[i] == 0:
            zero = nums.pop(i)
            nums.append(zero)
    return nums

print(moveZeros([0,1,0,3,12]))
print(moveZeros([0,0,1])) 
#In the above example code fails, 
# Action	              Result
# pop(i)	              shifts elements left
# for i in range(...)	  blindly moves to next index
# Outcome	              shifted element is skipped

def moveZeross(nums):
        insert_pos = 0
        for i in range(len(nums)):
            if nums[i]!=0:
                nums[insert_pos]=nums[i]
                insert_pos += 1
        for i in range(insert_pos, len(nums)):
             nums[i]=0
        return nums


print(moveZeross([0,1,0,3,12]))
print(moveZeross([0,0,1])) 
```

```python
# 5ï¸âƒ£ Merge Two Sorted Arrays
# Problem:
# Merge two sorted arrays into one sorted array (often in-place).
# Key idea: Compare from the end to avoid overwriting.
# Example
# nums1 = [1,2,3,0,0,0], nums2 = [2,5,6]
# Output â†’ [1,2,2,3,5,6]

#My approach
def mergeArray(nums1, nums2):
    pos = 0
    for i in range(len(nums1)):
        if nums1[i] > nums2[pos] and pos < len(nums2):
            nums1[i],nums2[pos]=nums2[pos],nums1[i]
            pos += 1
    #print(nums1)
    del(nums1[len(nums2):])
    #print(nums1)
    nums1 = nums1 + nums2

    return nums1
print(mergeArray([1,2,3,0,0,0],[2,5,6]))
```

```python
# 6ï¸âƒ£ Contains Duplicate
# Problem:
# Check if the array contains any duplicate.
# Key idea: Use a set.
# Example
# nums = [1,2,3,1]
# Output â†’ True

def contains_Duplicate(nums):
    no_duplicates = list(set(nums))
    return no_duplicates

print(contains_Duplicate([1,2,3,1]))
```

```python
# 9ï¸âƒ£ Valid Anagram
# Problem:
# Check if two strings contain the same characters with same frequency.
# Key idea: Frequency count / sorting.
# Example
# s = "anagram", t = "nagaram"
# Output â†’ True

def valid_anagram(string1, string2):
    return sorted(string1) ==  sorted(string2)

print(valid_anagram("ram","mars"))

#optimal
def valid_anagram(s, t):
    if len(s) != len(t):
        return False
    
    freq = {}

    for char in s:
        freq[char] = freq.get(char, 0)+1
    for char in t:
        if char not in freq:
            return False
        freq[char] -= 1
        if freq[char] == 0:
            del freq[char] #because if not 0 means {a:0}X means mismatch {a:2} so not anagram
    return True
print(valid_anagram('num','numm'))
print(valid_anagram('numm','numm'))
```

```python
# ðŸ”Ÿ Rotate Array
# Problem:
# Rotate the array to the right by k steps.
# Key idea: Reverse parts of the array.
# Example
# nums = [1,2,3,4,5,6,7], k = 3
# Output â†’ [5,6,7,1,2,3,4]

def rotateArray(nums, k):
    if k == 0 or k == len(nums):
        return nums
    elif k > len(nums):
        k = k%len(nums)
    return nums[len(nums)-k:] +nums[:len(nums)-k]

print(rotateArray([1,2,3,4,5,6,7],3))

#optimal
def rotateArray(nums,k):
    k %= len(nums)

    nums.reverse()
    nums[:k] = reversed(nums[:k])
    nums[k:] = reversed(nums[k:])
    return nums
print(rotateArray([1,2,3,4,5,6,7],3))
```
