```python
# #Problem:
# Given an array and a target, find indices of two numbers that add up to the target.
# Key idea: Use a hashmap to remember what you‚Äôve seen.
# Example
# nums = [2, 7, 11, 15], target = 9
# Output ‚Üí [0, 1]   (2 + 7 = 9)

#solution when you are checking in straignt line
def twosum(nums, target):
    add = 0
    for i in range(len(nums)-1):
        if target == nums[i] + nums[i+1]:
            return nums[i], nums[i+1]

#solution when you have to check back and forth
def twosums(nums, target):
    add = 0
    for i in range(len(nums)):
        for j in range(i+1, len(nums)-i): #i + 1 because i == j condition never occur(wrong sum)
            if target == nums[i] + nums[j]:
                return nums[i], nums[j]
print(twosum([2,7,11,15], 26))
print(twosums([2,7,11,15], 17))

#optimal solution

def twonumbersum(nums, target):
    seen = {}
    for idx, value in enumerate(nums):
        needed = target - value
        if needed in seen:
            return seen[needed], idx
        seen[value] = idx
print(twonumbersum([2,5,7,11,15],9)) # returns indexes


def twonumbersums(nums, target):
    seen = {}
    for idx, value in enumerate(nums):
        needed = target - value
        if needed in seen:
            return needed, value
        seen[value] = True
print(twonumbersums([2,5,7,11,15],9)) # returns values

```
